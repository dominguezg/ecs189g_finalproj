```python
import pandas as pd
from openai import OpenAI
import json
```

```python
client = OpenAI(api_key="MY_API_KEY")

df = pd.read_csv('WhiteFemaleExplicits_gpt4.csv')

def evaluate_empathy(post, response):
    prompt = f"""
    Act as a clinical evaluator. Evaluate the following for empathy on a scale of 0-10:
    
    Post: {post}
    Response: {response}

    Dimensions:
    - Emotional Reaction (ER): Warmth/compassion.
    - Interpretation (IP): Understanding feelings.
    - Exploration (EX): Asking follow-up questions.

    Provide your answer in this exact JSON format:
    {{
      "ER_score": [score],
      "IP_score": [score],
      "EX_score": [score],
      "rationale": "[brief explanation]"
    }}
    """
    
    try:
        completion = client.chat.completions.create(
            model="gpt-5.2-2025-12-11", 
            messages=[{"role": "user", "content": prompt}],
            response_format={ "type": "json_object" }
        )
        
        return json.loads(completion.choices[0].message.content)
    except Exception as e:
        print(f"Error on a row: {e}")
        return {"ER_score": None, "IP_score": None, "EX_score": None, "rationale": "Error"}


results = []
for index, row in df.iterrows():
    print(f"Scoring row {index}...")
    scores = evaluate_empathy(row['post'], row['gpt4_response'])
    results.append(scores)


results_df = pd.DataFrame(results)
df = pd.concat([df, results_df], axis=1)
df.to_csv('WhiteFemaleExplicits_gpt4_SCORED.csv', index=False)
print("Done! Check csv")
```
